```python
import pandas as pd

file_path = 'C:/INTERNSHIPS/Prodigy Infotech/Task 5/US_Accidents_March23.csv'
data = pd.read_csv(file_path)
print(data)
```

```python
missing_values=data.isnull().sum()
print(missing_values)
```

```python
mean_End_Lat=data['End_Lat'].mean()
data["End_Lat"].fillna(mean_End_Lat,inplace=True)
print(data)
```

```python
mean_End_Lng=data['End_Lng'].mean()
data["End_Lng"].fillna(mean_End_Lng,inplace=True)
print(data)
```
